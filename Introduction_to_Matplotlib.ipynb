### Pyplot API vs Object -Oriented (OO) API
* Pyplot API -> Quickly
* 00 API -> Advanced

#### Pyplot API

```python
x=np.linspace(0,10,1000)
plt.plot(x,np.sin(x), color='brown',linestyle="dashed",label="sin(x)");
plt.plot(x,np.cos(x), color='green', label="cos(x)");

plt.title("Sin(X) and Cos(X)")
plt.xlabel("Bien X")
plt.ylabel("Sin(X)")
plt.xlim([0,4])
plt.ylim([-1,1])

plt.legend();
```

#### Object-Oriented API

```python
#1. Prepare data
x=[1,2,3,4]
y=[11,22,33,44]

#2. Setup plot
fig, ax = plt.subplots(figsize=(5,5))

#3. Plot the data
ax.plot(x,y, color='red');
ax.set(title="A simple plot",
       xlabel="x-axis",
       ylabel="y-axis");
```

#### line

#### Scatter


```python
# Pyplot API
plt.scatter(x, np.exp(x), color='yellow'); #y=e^(x)
```

```python
# 00 API
# Prepare New Data
rng = np.random.RandomState(0)

x= rng.randn(100)
y= rng.randn(100)

colors = rng.rand(100)
sizes = 1000*rng.rand(100)
fig, ax= plt.subplots()
x1=ax.scatter(x,y, s=sizes, c=colors, cmap='viridis', alpha=0.3);

fig.colorbar(x1);
```

#### 3. Bar
* Vertical
* Horizontal

```python
sort_drink_prices = {"Coca": 10,
                     "Pepsi": 12,
                     "7up": 8}
fig, ax=plt.subplots()
ax.bar(sort_drink_prices.keys(), sort_drink_prices.values(), color='orange');
ax.set(title=" Bach Hoa Xanh", ylabel=" Price ($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(sort_drink_prices.keys()), list(sort_drink_prices.values()), color='green')
ax.set(title=" Bach Hoa Xanh", ylabel=" Price ($)");
```

#### 4. Histogram

```python
#Prepare data
np.random.seed(42)
student_height = np.random.normal(170,10, 250)
```

```python
student_height[:10], len(student_height)
```

```python
#Pyplot API
plt.hist(student_height, color='green', bins=30);
```

#### Subplots

```python
# Option 1:
fig, ((ax1,ax2),(ax3,ax4))= plt.subplots(nrows=2,
                                         ncols=2,
                                         figsize=(10,5))
#Plot data to each axis
ax1.plot(x,x/2)
ax1.plot(x,x*2)
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(sort_drink_prices.keys(), sort_drink_prices.values())
ax4.hist(student_height, bins=30);
```

```python
# Option 2:
fig, ax= plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,5))
#Plot data to each axis
ax[0,0].plot(x,x/2)
ax[0,0].plot(x,x*2)
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].bar(sort_drink_prices.keys(), sort_drink_prices.values(),color='purple')
ax[1,1].hist(student_height, bins=30);
```

### Project: Data Visualization- California Cities

```python
cities= pd.read_excel(r'C:\Users\DELLFULL\Desktop\Data\A1.xlsx')
cities.head()
```

```python
#Extract Latd = 
lat, lon= cities['latd'], cities['longd']
lon[:10]
population, area = cities['population_total'], cities['area_total_km2']
plt.style.use('seaborn')
plt.figure(figsize=(8,6));
```

```python
#Plot using Pyplot API
plt.scatter(lon, lat, 
            c=np.log10(population),cmap='viridis',
            s=area, linewidths=0, alpha=0.5)

plt.axis('equal')
plt.xlabel('longlatitude')
plt.ylabel('latitude')
plt.colorbar(label='log$_[10]$(population)')
plt.clim(3,7)

#Create a legend for cities'sizes
area_range =[50,100,300,500]
for area in area_range:
    plt.scatter([],[],s=area, c='k', alpha=0.4,
                label=str(area)+'km$^2$')
plt.legend(labelspacing=1, title='City Area')
plt.title('California Cities: Population and Area Distribution');
```
